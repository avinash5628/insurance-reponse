```python
na_df=pd.DataFrame({"columns_name":df.columns,
                    "Na_count":df.isnull().sum(),
                    "Na_percentage":df.isnull().sum()/df.shape[0]*100})
na_df.sort_values(by="Na_percentage",ascending=False)
```

```python
num_cols=[col for col in df.columns if df[col].dtype=="int64" or df[col].dtype=="float64"]
num_cols

for ele in num_cols:
    df[ele]=df[ele].fillna(df[ele].mean())
```

```python
cat_cols=[]
for col in df.columns:
    if df[col].dtype=="object":
        cat_cols.append(col)
cat_cols


for ele in cat_cols:
    df[ele]=df[ele].fillna(df[ele].mode()[0])
```

```python
df[num_cols].boxplot(figsize=(18,25),showmeans=True)
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
label=LabelEncoder()
Scaler=StandardScaler()
```

```python
for ele in cat_cols:
    df[ele]=label.fit_transform(df[ele]).astype(str)
```

```python
for ele in num_cols:
    df[ele]=Scaler.fit_transform(np.array(df[ele]).reshape(-1,1))
```

```python
df.describe()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
